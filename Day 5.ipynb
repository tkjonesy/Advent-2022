```python
file = open('input5.txt', 'r')
lines = file.readlines()
```

```python
horizontals = []
for x in range(9):
    horizontals.append([])
cleanlines = []
for line in lines:
    if '[' not in line:
        continue
    line = line.replace('[', ' ')
    line = line.replace(']', ' ')
    line = line.replace('\n', '')
    cleanlines.append(line)
for index, line in enumerate(cleanlines):
    stack = 0
    for i, char in enumerate(line):
        if char == ' ':
            for x, val in enumerate(cleanlines[index:]):
                if x == len(cleanlines) - 1:
                    continue
                if cleanlines[index + x][i] != ' ':
                    stack += 1
                    break
                else:
                    continue
        elif char == ' ':
            continue
        else:
            stack += 1
            height = abs(index - 8)
            horizontals[stack - 1].append(char)
            
print(horizontals)
```

```python
final = []
for i, lis in enumerate(horizontals):
    lis.reverse()
    final.append(lis)
print(final)
```

```python
for line in lines:
    if 'move' not in line:
        continue
        
    linesplit = line.split()
    source = int(linesplit[3])
    destination = int(linesplit[-1])
    times = int(linesplit[1])
    
    for x in range(times):
        final[destination - 1].append(final[source - 1][-1])
        final[source - 1].pop(-1)
```

```python
for i in final:
    print(i[-1])
```
