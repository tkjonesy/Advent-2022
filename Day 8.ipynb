```python
file = open('input8.txt', 'r')
lines = file.readlines()
```

```python
total_visible = 0

for index, line in enumerate(lines):
    line = line.replace('\n', '')
    if index == 0:
        total_visible += len(line)
        print(line, total_visible)
        continue
    elif index == len(lines) - 1:
        total_visible += len(line)
        print(line, total_visible)
        continue
    for i, height in enumerate(line):
        height = int(height)
        if i == 0:
            total_visible += 1
            print(f' Tree height: {height}, first tree, x: {i}, y: {index}, Total Visible: {total_visible}')
            continue
        elif i == len(line) - 1:
            total_visible += 1
            print(f' Tree height: {height}, last tree, x: {i}, y: {index}, Total Visible: {total_visible}')
            continue
        current_x = i
        current_y = index

        visible_left = False
        visible_right = False
        visible_up = False
        visible_down = False
        flag = False
        # Looking left
        while True:
            if current_x == 0:
                if visible_left:
                    total_visible += 1
                    print(f' Tree height: {height}, visible from left, x: {i}, y: {index}, Total Visible: {total_visible}')
                    flag = True
                    break
                else:
                    break
            elif height <= int(line[current_x - 1]):
                visible_left = False
                break
            else:
                visible_left = True
                current_x -= 1

        if flag:
            continue

        current_x = i
        current_y = index

        # Looking right
        while True:
            if current_x == len(line) - 1:
                if visible_right:
                    total_visible += 1
                    print(f' Tree height: {height}, visible from right, x: {i}, y: {index}, Total Visible: {total_visible}')
                    flag = True
                    break
                else:
                    break
            elif height <= int(line[current_x + 1]):
                visible_right = False
                break
            else:
                visible_right = True
                current_x += 1

        if flag:
            continue

        current_x = i
        current_y = index

        # Looking up
        while True:
            if current_y == 0:
                if visible_up:
                    total_visible += 1
                    print(f' Tree height: {height}, visible from top, x: {i}, y: {index}, Total Visible: {total_visible}')
                    flag = True
                    break
                else:
                    break
            elif height <= int(lines[current_y - 1][i]):
                visible_up = False
                break
            else:
                visible_up = True
                current_y -= 1

        if flag:
            continue

        current_x = i
        current_y = index

        # Looking down
        while True:
            if current_y == len(lines) - 1:
                if visible_down:
                    total_visible += 1
                    print(f' Tree height: {height}, visible from bottom, x: {i}, y: {index}, Total Visible: {total_visible}')
                    break
                else:
                    break
            elif height <= int(lines[current_y + 1][i]):
                visible_down = False
                break
            else:
                visible_down = True
                current_y += 1

print(total_visible)
```

```python
highest_score = 0

for index, line in enumerate(lines):
    line = line.replace('\n', '')
    if index == 0:
        continue
    elif index == len(lines) - 1:
        continue
    for i, height in enumerate(line):
        height = int(height)
        total = 0
        if i == 0:
            continue
        elif i == len(line) - 1:
            continue
        current_x = i
        current_y = index

        visible_left = 0
        visible_right = 0
        visible_up = 0
        visible_down = 0
        
        # Looking left
        while True:
            if current_x == 0:
                break
            elif height <= int(line[current_x - 1]):
                visible_left += 1
                break
            else:
                visible_left += 1
                current_x -= 1

        current_x = i
        current_y = index

        # Looking right
        while True:
            if current_x == len(line) - 1:
                break
            elif height <= int(line[current_x + 1]):
                visible_right += 1
                break
            else:
                visible_right += 1
                current_x += 1

        current_x = i
        current_y = index

        # Looking up
        while True:
            if current_y == 0:
                break
            elif height <= int(lines[current_y - 1][i]):
                visible_up += 1
                break
            else:
                visible_up += 1
                current_y -= 1

        current_x = i
        current_y = index

        # Looking down
        while True:
            if current_y == len(lines) - 1:
                break
            elif height <= int(lines[current_y + 1][i]):
                visible_down += 1
                break
            else:
                visible_down += 1
                current_y += 1

        total = visible_left * visible_right * visible_down * visible_up
        print(f'Tree: {height} Vision Score: ({visible_left} X {visible_right} X {visible_down} X {visible_up}) = {total}')
        if total > highest_score:
            highest_score = total
```
